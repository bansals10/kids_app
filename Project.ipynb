```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import json
```

```python
with open('online_safety.json') as json_file:
    data = json.load(json_file)

df = pd.DataFrame(data)
df.groupby('genre')['title'].nunique().plot(kind='bar')
```

```python
df.groupby('genreId')['title'].nunique().plot(kind='bar')
plt.show()
```

```python
df.groupby('familyGenre')['title'].nunique().plot(kind='bar')
```
